# Programs on string

```python
# find the length of givn string without len() functioon
s = input('enter a string')
count = 0

for i in s:
    count+=1
print(count)
```

```python
# Extract username from given email
s = input('enter a gmail')

s1 = s.index('@')

print(s[0:s1])
```

```python
# count the frequency of character
s = input('enter a line')
term = input("which character find you")

count = 0

for i in s:
    if i == term:
        count+=1
print(count)
    
```

```python
# write a progra wich can remove particular character from a string
s = input('enter a line')
term = input("which character remove you")

result = ''
for i in s:
    if i != term:
        result = result+i         # avoid remove char and add other char
print(result)
```

```python
#
s = input('enter a line')
l = []
temp = ''
for i in s:
    if i != ' ':
        temp = temp+i
    else:
        l.append(temp)
        temp = ''
l.append(temp)
print(l)
        
        
```

```python
# write a program to convert string a title case without title()
s = input('enter the string')

l = []
for i in s.split():
    l.append(i[0].upper() + i[1:].lower())
print(l)

print(" ".join(l))
```
