# Reference Variables
- Reference variables hold the objects
- We can create objects without reference variable as well
- An object can have multiple reference variables
- Assigning a new reference variable to an existing object does not create a new object

```python
#object without a refrenced
class person():
    def __init__(self):
        self.name = 'vaibhav'
        self.gender = 'male'
        

p = person()
q =p
print(id(p))
        
print(id(q))
        
```

```python
print(p.name)
print(q.name)

q.name = 'chetan'   # ek se change kiya to sabhi reffrence me change ho jayega

print(p.name)
print(q.name)
```

### Pass by refrence

```python
class person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender
        
        
        
#outside the classs function
def greet(person):
    print(id(person))
    print("my name is",person.name,'and i am',person.gender)
    

p = person('nitesh','male')
print(id(p))
greet(p)
```

```python
class person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender
        
        
        
#outside the classs function
def greet(person):
    print(id(person))
    person.name = 'jayesh'
    print(person.name)             #change name
    
    

p = person('nitesh','male')
print(id(p))
greet(p)
print(p.name)
```

```python
# private variable 
# __name      ---> pritvate

class person:
    def __init__(self):
        self.__name = 'vaibhav'
        
    def show(self):
        print(self.__name)
        
p = person()

# hoe to change private variable --> _classnam__varablename  | _person__name = 'jfvfw'
p._person__name = 'chetan'

p.show()   
        
        
        
```

```python
class person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender
        
        
p1 = person('vaibhav','male')
p2 = person("chetan",'female')
p3 = person("jayesh",'female')

l = [p1,p2,p3]
for i in l:
    print(i.name,i.gender)
```

# Static keyword

```python
# nedd for static vars
class Calc:
    @staticmethod
    def add(a, b):
        return a + b

res = Calc.add(2, 3)
print(res)
```

```python
class Check:
    @staticmethod
    def is_even(n):
        return n % 2 == 0

print(Check.is_even(10))
```
