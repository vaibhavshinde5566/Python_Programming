## Super Keyword

```python
class phone():
    def __init__(self,price,brand,camera):      # this not execute because this class inherited and frist execute child class contrucotr
        print('inside phone construcotr')
        self.price = price                # private variable
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print('buying a phone')
        
class smartphone(phone):
    def buy(self):
        print('Buying a smart phone')
        super().buy()                      # if i use super keyword then  parent class method print 
         
s = smartphone(2000,"apple",50)
s.buy()    
```

```python
# super ---> contrusctor
class phone():
    def __init__(self,price,brand,camera):      # this not execute because this class inherited and frist execute child class contrucotr
        print('inside phone construcotr')
        self.price = price                # private variable
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print('buying a phone')
        
class smartphone(phone):
    def __init__(self, price, brand, camera,os,ram):
        print('inside smarphone contructor')
        super().__init__(price, brand, camera)       # call the parent class contructor
        self.os = os
        self.ram = ram
        print('inside smartphone constructor')

        
         
s = smartphone(2000,"apple",50,'android',2)
print(s.os)
print(s.brand)

```

```python
# que
class parent():
    def __init__(self,num):
        self.__num =num
        
    def get_num(self):
        return self.__num
    
class child(parent):
    def __init__(self,val,num):
        super().__init__(num)
        self.__val = val
        
        
    def get_val(self):
        return self.__val
        
    

    
son = child(200,100)
print("parent",son.get_num())
print(son.get_val())
```

```python
# single
class phone():
    def __init__(self,price,brand,camera):      # this not execute because this class inherited and frist execute child class contrucotr
        print('inside phone construcotr')
        self.price = price                # private variable
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print('buying a phone')
        
class smartphone(phone):
    pass

smartphone(1000,"apple","13px").buy()
```

```python
# multilevel

class product:
    def review(self):
        print("product costomer review")
    
class phone(product):
    def __init__(self,price,brand,camera):      # this not execute because this class inherited and frist execute child class contrucotr
        print('inside phone construcotr')
        self.price = price                # private variable
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print('buying a phone')
        
class smartphone(phone):
    pass

s = smartphone(2000,"apple",12)
s.buy()
s.review()
```

```python
#hierarchical
class phone():
    def __init__(self,price,brand,camera):      # this not execute because this class inherited and frist execute child class contrucotr
        print('inside phone construcotr')
        self.price = price                # private variable
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print('buying a phone')
        
class smartphone(phone):
    pass
class featurephone(phone):
    pass


smartphone(100,'apple',50).buy()

featurephone(4000,'android',30).buy()
```

```python
# multiple

class phone():
    def __init__(self,price,brand,camera):      # this not execute because this class inherited and frist execute child class contrucotr
        print('inside phone construcotr')
        self.price = price                # private variable
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print('buying a phone')
        
class product(phone):
    def buy(self):
        print('product buy method')
    pass
class smartphone(product,phone):      # order matter 
    pass
    
    


s = smartphone(100,'apple',50)
s.buy()
```
