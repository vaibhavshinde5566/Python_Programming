## *args vs **kwargs
- *args and **kwargs are special python kewwords that are used to pass variable of argument to a function

```python
# *args
# # *args allow us to pass a variable number of kewords argument to a function

def multiply(*args):                #you can give any name like *vaibhav
    product = 1
    for i in args:
        product = product*i
    return product
multiply(1,2,3,5,3,5,4)      # you pass multiple input means arguments
```

```python
# **kwargs 
# **kwrags allows us to pass any number of kewrod argument
# keyword argument mean that they contain a key-value pair , like a python dictonary

def display(**kwargs):
    for (key,value) in kwargs.items():
        print(key,' -->',value)
display(india ='delhi',shrilanka = 'colombo',nepal = 'kathmandu')
```

## Nested Functions

```python
def f():
    def g():
        print("inside function g")
        f()
    g()
    print("inside if f")
f()

# this is created looped
```

```python
def g(x):
    def h(x):
        x = x+1
        print("in h(x): x = ",x)
    x = x+1
    print('in g(x): x = ',x)
    h(x)
    return x

x = 3
z = g(x)
print('in main program scope : x = ',x)
print('in main program scope : z = ',z)
```

# Lambda Function
- lambada is a anonou

```python
# x->x^2
a = lambda x:x**2
a(2)
```

```python
# x,y -->x^2
a = lambda x,y:x+y
a(5,2)
```

```python
# cheack string has 'a'
a = lambda s:'a' in s
a('helalo')
```

```python
# odd or even
a = lambda b:'even' if b%2==0 else 'odd'
a(3)
```

## High order function

```python
def square(x):
    return x**2

def transform(f,l):     # f is call to square function
    output = []
    for i in l:
        output.append(f(i))
    print(output)
    
l = [1,2,3,4,5]

transform(square,l)
    
```

# Map

```python
# square the items of list
l = [1,2,3,4,5]
list(map(lambda x:x**2,l))
```

```python
# odd /even labelling
l = [1,2,3,4,5]
list(map(lambda x:'even' if x%2==0 else 'odd',l))
```

# Filter

```python
# fetch fruits starting with 'a'
fruits = ['cherry','apple','banaa']
list(filter(lambda x:x.startswith('a'),fruits))
```

# Reduce
