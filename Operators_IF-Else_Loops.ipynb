```python
# print th table 
number = int(input('enter number'))

i = 1
while i<11:
    print(number*i)
    i+=1
```

```python
# while loop with else
x =1
while x<5:
    print(x)
    x+=1
else:
    print('liit cross')
```

```python
#guessing game

import random
jackpot = random.randint(1,100)

guess = int(input('guess number'))

while guess != jackpot:
    if guess < jackpot:
        print('wrong!guess higher')
    else:
         print('wrong!guess lower')
    guess = int(input('guess number'))
         
```

```python
#pattern
n = 5
for i in range(0,n):
    for j in range(i+1):
        print('*',end='')
    print()
    
```

```python
#pattern 2  numbwrs
n = 5
for i in range(0,n+1):
    for j in range(1,i+1):
        print(j,end='')
    for k in range(i-1,0,-1):
        print(k,end='')
    print()
```

# Loop control
- Break
- Continue
- Pass

```python
for i in range(1,10):
    if i == 5:
        break          #stop loop when cond is met
    print(i)
```

```python
for i in range(1,5):
    if i==3:
        continue     # skip 3 and continue loop
    print(i)
        
```

```python
for i in range(10,20+1):
    for j in range(2,i):
        if i%2==0:
            break
        else:
            print(i)
           
        
```

```python
# Pass
for i in range(1,10):
    pass                    # with pass not throw error
    
```
