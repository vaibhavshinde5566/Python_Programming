# Exception Handling
There are 2 stages where error may happen in a program

During compilation -> Syntax Error
During execution -> Exceptions

### Syntax Error

- Error is raised by the interpreter/compiler
- You can solve it by rectifying the program

```python
# Examples of syntax error
print 'hello world'
```

## Exceptions
- If things go wrong during the execution of the program(runtime). It generally happens when something - - unforeseen has happened.

- Exceptions are raised by python runtime
You have to takle is on the fly
## Examples
- Memory overflow
- Divide by 0 -> logical error
- Database error

```python
try:
    with open('saple.txt','r') as f:
        print(f.read())
except:
    print('sory file notfound')    #when file not foound
```

```python
# Cathching specific exception
try:
    m=5
    f= open('sample.txt','r')
    print(f.read())
    print(m)
    print(5/2)
    l = [24,2,4,2,3]
    l[100]
    
except FileNotFoundError:
    print('file not found')
except NameError:
    print('variable not defined')
except ZeroDivisionError:
    print('cannot devide by zero')
except Exception as e:
    print(e)
    
```

```python
# ELSE
try:
    f = open('sample.txt','r')     # if file 
except FileNotFoundError:
    print('file nahi hai')
except Exception as e:
    print(e)
else:
    print(f.read())                  # execute when file is correct 
```

```python
# Finally

try:
    f = open('sample.txt','r')     # if file 
except FileNotFoundError:
    print('file nahi hai')
except Exception as e:
    print(e)
else:
    print(f.read())  # execute when file is correct 
finally:
    print('ye print hoga hi')
```

```python
class bank:
    def __init__(self,balance):
        self.balance = balance
        
    def withdraw(self,amount):
        if amount<0:
            raise Exception('amount cannot be negative')
        if self.balance < amount:
            raise Exception('paisa nahi tere pass')
        self.balance = self.balance-amount
        
obj = bank(10000)
try:
    obj.withdraw(-5000)
except Exception as e:
    print(e)
else:
    print(obj.balance)
```
