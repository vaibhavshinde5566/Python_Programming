# Using Context Manager (With)
- It's a good idea to close a file after usage as it will free up the resources
- If we dont close it, garbage collector would close it
- with keyword closes the file as soon as the usage is over

```python
# with    -----> replacement of close 

with open('sample.txt','w') as f:
    f.write('saleman bhoi')
```

```python
# try f.read() 
with open('sample.txt','r') as f:
    print(f.read())
```

```python
# moving within a file --> 10 char then 10 char
with open('sample.txt','r') as f:
    print(f.read(10))
    print(f.read(10))
```

- seek and tell fuction

```python
with open('sample.txt','r') as f:
    print(f.read(10))
    print(f.tell())
    
    f.seek(0)         # start from frist
    print(f.read(10))
    print(f.tell())
```

```python
# seek during write
with open('sample.txt','w') as f:
    f.write('hello')
    f.seek(0)          # 0 position
    f.write('x')       # replace h ---> x  
```

# Serilization using json module
#Serialization and Deserialization
- Serialization - process of converting python data types to JSON format
- Deserialization - process of converting JSON to python data types

```python
# serialization

import json
l = [1,2,4,3]

with open('demo.json','w') as f:
    json.dump(l,f)                #l--> kyA dump karna hai   f-----> file handler
    
    
```

```python
# dict
d = {
    'name': 'vaibhav',
    'age':22,
    'gender': 'male'
}

with open('demo.json','w') as f:
    json.dump(d,f,indent=4)
```

```python
# Deserialization     reverse process of seralization

import json
with open('demo.json','r') as f:
    d = json.load(f)
print(d)
print(type(d))
    
```

```python
# serialize and deseralize with tuple

t = (1,2,4,5,3)
import json
with open('demo.json','w') as f:
    json.dump(t,f)
```

## Serialize and desealization costom  object


```python
class person():
    def __init__(self,fname,lname,age,gender):
        self.fname = fname
        self.lname = lname
        self.age = age
        self.gender = gender
p = person('vaiibhav','shhinde',22,'male')        
import json

## 
def show_obj(obj):
    if isinstance(obj,person):
        return "{} {} age {}  gender {}".format(obj.fname,obj.lname,obj.age,obj.gender)

with open('demo.json','w') as f:
    json.dump(p,f,default=show_obj)  
```

```python
# with dictonary
import json

def show_object(obj):
  if isinstance(obj,person):
    return {'name':obj.fname + ' ' + obj.lname,'age':obj.age,'gender':obj.gender}

with open('demo.json','w') as f:
  json.dump(p,f,default=show_object,indent=4)
```

# Pickling
Pickling is the process whereby a Python object hierarchy is converted into a byte stream, and unpickling is the inverse operation, whereby a byte stream (from a binary file or bytes-like object) is converted back into an object hierarchy.

```python
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
    def display(self):
        print('hi name is',self.name,'and i am ',self.age,'yearl old')
```

```python
# pickle dump
p = person('vaibhav',33)

import pickle
with open('person.pkl','wb') as f:
    pickle.dump(p,f)
```

```python
# pickle load
import pickle
with open('person.pkl','rb') as f:
    p = pickle.load(f)
    
p.display()
```
