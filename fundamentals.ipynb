```python
def is_prime(num):
    if num<=0:
        return False
    
    for i in range(2,num-1):
        if num%i==0:
            return False
    return True
num = 11
result = is_prime(num)
print('is {num} a prime number',result)

        
```

```python
import pandas as pandas
import numpy as np
marks = [45,76,89,53,67,90,82,49,33,99]


#filter student who passed
for m in marks:
    if m > 50:
        print("student Passed")
    else:
        print('student failed')
    
```

```python
# calculate grades
for m in marks:
    if m > 85:
        print(m,': A')
    elif  m>70 and m<84:
        print(m,': B ')
    elif m>50 and m<69:
        print(m,': C')
    else:
        print(m,': Fails')
```

```python
# add 5 grace marks who failed
n = 5
for m in marks:
    if m<50:
        m += n
    print(m)
```

```python
# avag of class
mean = np.mean(marks)
print('average of classs :',mean)
```

```python
class BankAccount:
    
    def __init__(self, balance):
        self.balance = balance
    
    def withdraw(self, amount):
        self.balance -= amount
        return self.balance
    
    def check_balance(self):
        return self.balance
    
    def month_end_check(self):
        if self.balance < 500:
            return "Please deposit money"
        else:
            self.balance -= 50
            return f"50 rupees charged. Final balance: {self.balance}"


# Object creation
account = BankAccount(1000)

# Withdraw 600
account.withdraw(600)

# Display balance
print("Total Balance:", account.check_balance())

# Month end check
print(account.month_end_check())

        
balance = 1000
        
```

```python
import pandas as pd

data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Product': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B'],
    'Sales': [200, 300, 150, 400, 250, 100, 300, 500],
    'Transactions': [2, 4, 3, 5, 1, 2, 6, 3]
    }

df = pd.DataFrame(data)

# total sales per region
total_sales = df.groupby('Region')['Sales'].sum()
total_sales
```

```python
#count of transaction per region
tran_count = df.groupby('Region')['Transactions'].count()
tran_count
```

```python
# avarage sales per region
avg_sales = df.groupby('Region')['Sales'].mean()
avg_sales
```

```python
# total number of transaction per region
total = df.groupby('Region')['Transactions'].sum()
total
```
